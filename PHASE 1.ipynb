```python
import serial
from firebase import firebase
import re

arduino = serial.Serial("COM5")
arduino.baudrate = 9600
sno = 0
temp = str(98)
base = firebase.FirebaseApplication('https://final-1e929.firebaseio.com/', None)

while(True):
        sno += 1
        xdata = arduino.readline()
        xdata = xdata.strip()
        xdata = str(xdata)
        xdata = re.sub('b','',xdata)

        data = {
        'sno' : sno,
        'val' : xdata,
        'temp' : temp
        }

        print(str(sno) + " " + xdata + " " + temp)

        base.post('/final-1e929/sensordata', data)
```
